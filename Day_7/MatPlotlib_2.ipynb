```python
#plt.subplot(nrows,ncols,index)
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,15,20,25]
plt.subplot(1,2,1)  #1st row,2nd colm, 1st index
plt.plot(x,y)
plt.title('Line Plot')

plt.subplot(1,2,2)   #1st row,2nd colm, 2st index
plt.bar(x,y)
plt.title('Bar Plot')
plt.tight_layout()
plt.show()
```

```python
#fig,ax=plt.subplots(nrows,ncolms,figsize=(width,hight))
fig,ax=plt.subplots(1,2,figsize=(12,4))
x=[1,2,3,4]
y=[10,15,20,25]

ax[0].plot(x,y)
ax[0].set_title('Line Plot')

ax[1].bar(x,y)
ax[1].set_title('Bar Plot')

plt.show()
```
