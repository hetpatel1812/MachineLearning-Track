```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.random.rand(100) * 10  # 100 values between 0 and 10
noise = np.random.randn(100) * 2  # random noise
y = 2 * x + 3 + noise  # true relation with noise
```

```python
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)
```

```python
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / n

print(f"ðŸ“ˆ Slope (m): {m:.2f}")
print(f"ðŸ“ˆ Intercept (c): {c:.2f}")
```

```python
y_pred = m * x + c
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="Data Points", color='skyblue')
plt.plot(x, y_pred, color='red', label="Regression Line")
plt.title("Linear Regression (from scratch)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
