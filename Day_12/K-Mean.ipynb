```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = {
    'Customer' : ['John', 'Jane', 'Doe', 'Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age' : [25, 34, 28, 45, 23, 35, 40, 29],
    'SpendingScore' : [500, 600, 550, 700, 450, 620, 680, 590]}
df = pd.DataFrame(data)
X =df[['Age','SpendingScore']]
```

```python
model = KMeans(n_clusters=2,random_state=42,n_init=10)
df['Group']= model.fit_predict(X)
plt.figure(figsize=(6,5))
for group in df['Group'].unique() :
    group_data = df[df['Group'] == group]
    plt.scatter(group_data['Age'], group_data['SpendingScore'], label=f'Group {group}')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation based on Age and Spending Score')
plt.legend()
plt.grid()
plt.show()
print(df)
```
