```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,15,25]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Mon','Tue','Wed','Thu','Fri']
y=[10,15,7,20,12]
plt.plot(x,y)
plt.title("Bakery sales In Week")
plt.xlabel('Day Of week')
plt.ylabel('Sales Of Cake')
plt.show()
```

```python
import matplotlib.pyplot as plt
Months=[1,2,3,4]
Sales=[1000,4000,5000,3000]
plt.plot(Months,Sales,color="blue",linestyle="--",marker='o',linewidth=2,label='2025 sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Data')
plt.legend(fontsize=12 )
plt.grid(color="gray")
plt.xlim(0,20)
plt.ylim(0,6000)
plt.xticks([1,2,3,4],['Jan','Feb','Mar','Apr'])
plt.show()
```

```python
import matplotlib.pyplot as plt
product=['A','B','C','D']
sales=[1000,2500,3100,1000]
plt.bar(product,sales,color="green",label="2025")
plt.xlabel("sales")
plt.ylabel("product")
plt.title("Product sales")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#plt.pie(vales,label="",colors="",autopc="%1.1f%%")
regions=["North","South","East","West"]
revenue=[3000,2000,4000,2000]
plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=['red','yellow','green','blue'])
plt.title('Revenue Contribution')
plt.show()
```
